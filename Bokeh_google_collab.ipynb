```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers

output_file('test.html')#render in html file but in output in console

```

```python
p=figure(title="Test flower")
p.xaxis.axis_label="patal lenght"
p.yaxis.axis_label="patal width"
p.circle(x=flowers['petal_length'], y=flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers

output_file('line.html')
p=figure(title="Test flower")
p.xaxis.axis_label="xaxis"
p.yaxis.axis_label="yaxis"
p.line(x=flowers['petal_length'], y=flowers['petal_width'])#circle=scatter
show(p)
```

```python
x=[1,2,3,4,5,6]
y=[5,6,7,8,9,10]
output_file('line.html')
p=figure(title="Line plot")
p.line(x,y)
show(p)
```

```python
x=[1,2,3,4,5,6]
y=[5,6,7,8,9,10]
output_file('line.html')
p=figure(title="Line plot")
p.scatter(x,y)
show(p)
```

```python
x=[1,2,3,4,5,6]
y=[5,6,7,8,9,10]
output_file('line.html')
p=figure(title="Line plot")
p.scatter(x,y,fill_color="red",legend_label='red points',size=20)
show(p)
```
